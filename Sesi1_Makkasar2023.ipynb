```python
import pandas as pd

df = pd.read_csv("/content/delaney.csv")

df
```

```python
y = df['logS']
y
```

```python
X = df.drop('logS', axis=1)
X
```

```python
# membagi data training dan data tes
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.2, random_state=100)
```

```python
X_test
```

```python
y_train
```

```python
# membuat dan mentraining model - Linier REgression

from sklearn.linear_model import LinearRegression

lr = LinearRegression()
lr.fit(X_train, y_train)
```

```python
# membuat prediksi dengan model

y_lr_train_pred = lr.predict(X_train)
y_lr_test_pred = lr.predict(X_test)
```

```python
# evaluasi model linier regression | lr
from sklearn.metrics import mean_squared_error, r2_score

lr_train_mse = mean_squared_error(y_train, y_lr_train_pred)
lr_train_r2 = r2_score(y_train, y_lr_train_pred)

lr_test_mse = mean_squared_error(y_test, y_lr_test_pred)
lr_test_r2 = r2_score(y_test, y_lr_test_pred)
```

```python
lr_train_mse
```

```python
lr_train_r2
```

```python
lr_test_mse
```

```python
lr_test_r2
```

```python
print('LR MSE Train:', lr_train_mse)
print('LR R2 Train:', lr_train_r2)
print('LR MSE Test:', lr_test_mse)
print('LR R2 Test:', lr_test_r2)
```

```python
lr_result = pd.DataFrame(['Linear regression',lr_train_mse, lr_train_r2, lr_test_mse, lr_test_r2 ]).transpose()
lr_result.columns = ['Mothod','Training MSE','Training R2', 'Test MSE','Test R2']
```

```python
lr_result
```

```python
# Random Forest Model
from sklearn.ensemble import RandomForestRegressor

rf = RandomForestRegressor(max_depth=2, random_state=100)
rf.fit(X_train, y_train)
```

```python
y_rf_train_pred = rf.predict(X_train)
y_rf_test_pred = rf.predict(X_test)
```

```python
# evaluasi model random Forest
from sklearn.metrics import mean_squared_error, r2_score

rf_train_mse = mean_squared_error(y_train, y_rf_train_pred)
rf_train_r2 = r2_score(y_train, y_rf_train_pred)

rf_test_mse = mean_squared_error(y_test, y_rf_test_pred)
rf_test_r2 = r2_score(y_test, y_rf_test_pred)
```

```python
rf_result = pd.DataFrame(['RandomForest',rf_train_mse, rf_train_r2, rf_test_mse, rf_test_r2 ]).transpose()

rf_result.columns = ['Mothod','Training MSE','Training R2', 'Test MSE','Test R2']
rf_result
```

```python
# combine result

rf_result
```

```python
df_models = pd.concat([lr_result, rf_result], axis=0)
df_models
```
