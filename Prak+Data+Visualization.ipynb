# Pengenalan Data Visualization

```python
import numpy as np
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
import seaborn as sns
```

## Ticks

```python
plt.figure(figsize=(6, 3))
plt.plot([2, 3 ,5], 'o')
plt.plot([3, 2, 4], 's')
plt.xticks(ticks=np.arange(3), labels=['Januari', 'Februari', 'Maret'], rotation=30)
plt.show()
```

## Legends

```python
plt.figure()
plt.plot([1, 2, 5], label='Satu')
plt.plot([2, 3, 4], label='Dua')
plt.legend()
plt.show()
```

# Bar Chart

```python
plt.figure()
plt.bar(['Kota A', 'Kota B', 'Kota C', 'Kota D'], [20, 35, 55, 30])
plt.show()
```

```python
plt.figure()
labels = ['Kota A', 'Kota B', 'Kota C', 'Kota D']
x = np.arange(len(labels))
width = 0.4
plt.bar(x - width / 2, [20, 35, 55, 30], width=width)
plt.bar(x + width / 2, [30, 15, 45, 30], width=width)
plt.xticks(x, labels)
plt.show()
```

# Pie Chart

```python
plt.figure()
plt.pie([10, 30, 40, 20], labels=['A', 'B', 'C', 'D'], explode=(0, 0, 0.1, 0), colors=['green', 'orange', 'blue', 'yellow'])
plt.show()
```

```python
plt.figure()
labels = ['A', 'B', 'C']
x = np.arange(len(labels))
bar1 = [5, 15, 20]
bar2 = [10, 15, 25]
bar3 = [10, 25, 10]
plt.bar(x, bar1)
plt.bar(x, bar2, bottom=bar1)
plt.bar(x, bar3, bottom=np.add(bar1, bar2))
plt.xticks(x, labels)
plt.show()
```

# Stacked Area Chart

```python
plt.figure()
plt.stackplot([1, 2, 5, 7], [3, 5, 9, 7], [5, 8, 7,10], [2, 4, 6, 8])
plt.show()
```

# Histogram

```python
plt.figure()
x = np.random.normal(0, 1, size=100)
plt.hist(x, bins=20, density=True)
plt.show()
```

# Scatter Plot

```python
x = np.random.normal(0,1, size=100)
y = x + np.random.normal(0, 1, size=100)
plt.figure()
plt.scatter(x, y)
plt.show()
```
