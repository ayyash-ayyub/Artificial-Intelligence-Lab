```python
# baca data
df = pd.read_csv("/content/heart.data.csv")
df.head(10)
```

```python
df = df.drop("Unnamed: 0", axis=1)
df
```

```python
sns.lmplot(x='biking', y='heart.disease', data=df)
sns.lmplot(x='smoking', y='heart.disease', data=df)
```

```python
x_df = df.drop('heart.disease', axis=1)
y_df = df['heart.disease']
```

```python
y_df
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(x_df, y_df, test_size=0.3, random_state=45)
```

```python
# buat model
from sklearn import linear_model
```

```python
model = linear_model.LinearRegression()
```

```python
model.fit(X_train, y_train)
```

```python
print(model.score(X_train, y_train))
```

```python
# generate model
import pickle
pickle.dump(model, open('model_siang.pkl', 'wb'))
```

```python
model = pickle.load(open('model_siang.pkl', 'rb'))
print(model.predict([[44.2,2.0]]))
```
