```python
df = pd.read_csv('/content/dataset_pegawai_sesi3.csv')
df.head()
```

```python
# Menghitung nilai mean, median, dan modus untuk kolom 'masa_kerja' dan 'penghasilan_rupiah'
mean_masa_kerja = df['masa_kerja'].mean()
median_masa_kerja = df['masa_kerja'].median()
modus_masa_kerja = df['masa_kerja'].mode()[0]
```

```python
print("Masa Kerja (tahun) - Mean:", mean_masa_kerja)
print("Masa Kerja (tahun) - Median:", median_masa_kerja)
print("Masa Kerja (tahun) - Modus:", modus_masa_kerja)
```

```python
mean_penghasilan = df['penghasilan_rupiah'].mean()
median_penghasilan = df['penghasilan_rupiah'].median()
modus_penghasilan = df['penghasilan_rupiah'].mode()[0]
```

```python
print("Penghasilan (Rupiah) - Mean:", mean_penghasilan)
print("Penghasilan (Rupiah) - Median:", median_penghasilan)
print("Penghasilan (Rupiah) - Modus:", modus_penghasilan)
```

```python
import matplotlib.pyplot as plt
```

```python
# 1. Grafik Masa Kerja
plt.figure(figsize=(10, 5))
plt.hist(df['masa_kerja'], bins=15, color='skyblue', edgecolor='black', alpha=0.7)
plt.axvline(mean_masa_kerja, color='red', linestyle='dashed', linewidth=2, label=f'Mean: {mean_masa_kerja:.2f}')
plt.axvline(median_masa_kerja, color='green', linestyle='dashed', linewidth=2, label=f'Median: {median_masa_kerja}')
plt.axvline(modus_masa_kerja, color='blue', linestyle='dashed', linewidth=2, label=f'Modus: {modus_masa_kerja}')
plt.title('Distribusi Masa Kerja Pegawai')
plt.xlabel('Masa Kerja (tahun)')
plt.ylabel('Jumlah Pegawai')
plt.legend()
plt.show()
```

```python
# 2. Grafik Penghasilan
plt.figure(figsize=(10, 5))
plt.hist(df['penghasilan_rupiah'], bins=15, color='lightgreen', edgecolor='black', alpha=0.7)
plt.axvline(mean_penghasilan, color='red', linestyle='dashed', linewidth=2, label=f'Mean: {mean_penghasilan:.2f}')
plt.axvline(median_penghasilan, color='green', linestyle='dashed', linewidth=2, label=f'Median: {median_penghasilan}')
plt.axvline(modus_penghasilan, color='blue', linestyle='dashed', linewidth=2, label=f'Modus: {modus_penghasilan}')
plt.title('Distribusi Penghasilan Pegawai')
plt.xlabel('Penghasilan (Rupiah)')
plt.ylabel('Jumlah Pegawai')
plt.legend()
plt.show()
```

menghitung nilai skewness dari masing-masing kolom numerik (masa_kerja dan penghasilan_rupiah) serta membuat grafik distribusinya untuk memvisualisasikan apakah datanya miring ke kiri (negatif), simetris, atau miring ke kanan (positif).

```python
from scipy.stats import skew
```

```python
# Hitung skewness untuk setiap kolom numerik
skewness_masa_kerja = skew(df['masa_kerja'])
skewness_penghasilan = skew(df['penghasilan_rupiah'])

print(f"Skewness Masa Kerja: {skewness_masa_kerja:.2f}")
print(f"Skewness Penghasilan: {skewness_penghasilan:.2f}")

# Plot histogram untuk visualisasi distribusi masa kerja
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
plt.hist(df['masa_kerja'], bins=15, color='skyblue', edgecolor='black', alpha=0.7)
plt.axvline(df['masa_kerja'].mean(), color='red', linestyle='dashed', linewidth=2, label=f'Mean')
plt.axvline(df['masa_kerja'].median(), color='green', linestyle='dashed', linewidth=2, label=f'Median')
plt.title(f'Distribusi Masa Kerja (Skewness: {skewness_masa_kerja:.2f})')
plt.xlabel('Masa Kerja (tahun)')
plt.ylabel('Jumlah Pegawai')
plt.legend()

# Plot histogram untuk visualisasi distribusi penghasilan
plt.subplot(1, 2, 2)
plt.hist(df['penghasilan_rupiah'], bins=15, color='lightgreen', edgecolor='black', alpha=0.7)
plt.axvline(df['penghasilan_rupiah'].mean(), color='red', linestyle='dashed', linewidth=2, label=f'Mean')
plt.axvline(df['penghasilan_rupiah'].median(), color='green', linestyle='dashed', linewidth=2, label=f'Median')
plt.title(f'Distribusi Penghasilan (Skewness: {skewness_penghasilan:.2f})')
plt.xlabel('Penghasilan (Rupiah)')
plt.ylabel('Jumlah Pegawai')
plt.legend()

plt.tight_layout()
plt.show()
```

kita dapat menampilkan sebaran penghasilan berdasarkan jenis jabatan, untuk ini dapat menggunakan boxplot atau violin plot. Kedua jenis plot ini akan menunjukkan distribusi penghasilan untuk setiap kategori jabatan (manager, direktur, dan staff biasa), sehingga lebih mudah untuk melihat rentang, kuartil, serta outlier dari penghasilan per jenis jabatan.

```python
import seaborn as sns
```

```python
# Plot menggunakan boxplot
plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x='jenis_jabatan', y='penghasilan_rupiah', palette="Set3")
plt.title('Sebaran Penghasilan Berdasarkan Jenis Jabatan')
plt.xlabel('Jenis Jabatan')
plt.ylabel('Penghasilan (Rupiah)')
plt.show()
```

keterangan:
Boxplot: Memvisualisasikan sebaran data dengan menunjukkan median, kuartil, serta potensi outlier di masing-masing kategori jenis_jabatan.
Jenis Jabatan: Kategori pada sumbu x (manager, direktur, staff biasa), dan nilai penghasilan pada sumbu y.

alternatif plot dengan: Violin Plot

```python
plt.figure(figsize=(10, 6))
sns.violinplot(data=df, x='jenis_jabatan', y='penghasilan_rupiah', palette="Pastel1")
plt.title('Distribusi Penghasilan Berdasarkan Jenis Jabatan')
plt.xlabel('Jenis Jabatan')
plt.ylabel('Penghasilan (Rupiah)')
plt.show()
```

contoh lain generalisasi analisis deskriptif dari data diatas

```python
# 1. Boxplot untuk Penghasilan Berdasarkan Jenis Jabatan
plt.figure(figsize=(12, 6))
sns.boxplot(data=df, x='jenis_jabatan', y='penghasilan_rupiah', palette="Set2")
plt.title('Distribusi Penghasilan Berdasarkan Jenis Jabatan')
plt.xlabel('Jenis Jabatan')
plt.ylabel('Penghasilan (Rupiah)')
plt.show()

# 2. Histogram dan KDE Plot untuk Masa Kerja
plt.figure(figsize=(12, 6))
sns.histplot(df['masa_kerja'], kde=True, color='skyblue', bins=15)
plt.title('Distribusi Masa Kerja Pegawai')
plt.xlabel('Masa Kerja (tahun)')
plt.ylabel('Jumlah Pegawai')
plt.show()

# 3. Boxplot untuk Masa Kerja Berdasarkan Jenis Jabatan
plt.figure(figsize=(12, 6))
sns.boxplot(data=df, x='jenis_jabatan', y='masa_kerja', palette="Set3")
plt.title('Distribusi Masa Kerja Berdasarkan Jenis Jabatan')
plt.xlabel('Jenis Jabatan')
plt.ylabel('Masa Kerja (tahun)')
plt.show()

# 4. Heatmap Korelasi
# Hanya dilakukan jika terdapat lebih dari dua kolom numerik dalam dataset
plt.figure(figsize=(8, 6))
sns.heatmap(df.corr(), annot=True, cmap="Blues", square=True)
plt.title('Matriks Korelasi Variabel Numerik')
plt.show()
```
