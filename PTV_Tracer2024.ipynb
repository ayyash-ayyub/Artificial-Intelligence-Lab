```python
import pandas as pd


file_name = '/content/data_ PTPPV.xlsx'

df = pd.read_excel(file_name, sheet_name='Sheet1')

print(df.head(10))
```

```python
# filter data untuk Tahun Lulus 2023
df_2023 = df[df['Tahun Lulus'] == 2023]

#  NIM yang duplikat
duplicate_nims = df_2023[df_2023.duplicated('NIM', keep=False)]

# group berdasarkan Nama PT dan  jumlah NIM yang duplikat
result = duplicate_nims.groupby('Nama PT')['NIM'].size().reset_index(name='Count')

#  hasil
print(result)
```
